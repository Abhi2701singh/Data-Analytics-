Bar Plot in seaborn lib

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
df1 = sns.load_dataset('penguins')
```

```python
df1
```

```python
df1["island"]
```

```python
sns.barplot(x=df1.island , y=df1.bill_length_mm)
plt.show()
```

###               ( OR )

```python
sns.barplot(x='island' , y='bill_length_mm' , data=df1)

plt.show()
```

```python
sns.barplot(x='island' , y='bill_length_mm' , data=df1, hue='sex')

plt.show()
```

# order 

```python
order1=["Dream" , "Torgersen","Biscoe"]

sns.barplot(x='island', y='bill_length_mm', data=df1,hue='sex',order=order1)
plt.show()
```

# hue_order

```python
order1=["Dream" , "Torgersen","Biscoe"]

sns.barplot(x='island', y='bill_length_mm', data=df1,hue='sex',order=order1, hue_order=['Female' , 'Male'])
plt.show()
```

# palette       ( graph ke colors set karne ke liye hota hai (jaise barplot me different hues ke liye alag colors dena). )

```python
order1=['Dream','Torgersen','Biscoe']
sns.barplot(x='island', y='bill_length_mm' , data = df1 , order=order1 , hue='sex', hue_order=['Female','Male'] ,  palette="Accent")
plt.show()
```

```python
order1=['Dream','Torgersen','Biscoe']
sns.barplot(x='island', y='bill_length_mm' , data = df1 , order=order1 , hue='sex', hue_order=['Female','Male'] ,  palette="prism")
plt.show()
```

```python
order1=['Dream','Torgersen','Biscoe']
sns.barplot(x='island', y='bill_length_mm' , data = df1 , order=order1 , hue='sex', hue_order=['Female','Male'] ,  palette="gist_earth")
plt.show()
```

# Saturation

```python
order1=['Dream','Torgersen','Biscoe']
sns.barplot(x='island', y='bill_length_mm' , data = df1 , order=order1 , hue='sex', hue_order=['Female','Male'] ,  palette="Accent" , saturation=0.2)
plt.show()
```

## Histogram in seaborn lib.

```python
sns.displot(df1["flipper_length_mm"])
plt.show()
```

# bins

```python
sns.displot(df1['flipper_length_mm'],bins=[170,180,190,200,210,220,230])
plt.show()
```

# color

```python
sns.displot(df1['flipper_length_mm'],bins=[170,180,190,200,210,220,230] , color='g')
plt.show()
```

# Kde

```python
sns.displot(df1['flipper_length_mm'] , bins=[170,180,190,200,210,220,230], color='r', kde='True')
plt.show()
```
