### Matplotlib

Matplotlib is a multi-platfrom data visualization lib. built on NumPy arrays and designed to work with the broader SciPy stack. It was introduce by John Hunter in the year 2002. One of the greatest benifits of visualization is that it allows us visual access to huge amount of data in easily diestible visuals.

1- Linear Plot

2- Scatter plot

3- Bar plot

4- Stem plot

5- Step plot

6- Hist plot

7- Box plot

8- Pie plot

9- Fill_between_plot

----------------------------------------------------------------------------------------------------------

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Linear graph

```python
x=[1,2,3,4]
y=[9,6,8,7]

plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[9,6,8,7]
c='r' # red color  

plt.plot(x,y,c)
plt.show()
```

```python
x=[1,2,3,4]
y=[9,6,8,7]

plt.plot(x,y, marker="o")
plt.show()
```

```python
x=[1,8,3,4,7]
y=[9,6,1,7,5]

xx='r'
yy='b'
plt.plot(x,xx,marker="o")
plt.plot(y,yy,marker=">")
plt.show()
```

## Bar graph

```python
x=[1,2,3,4]
y=[9,6,8,7]

plt.bar(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[9,6,8,7]
c=['r','g','y','b']
plt.bar(x,y,color=c)
plt.show()
```

```python
x= np.array(["A","B","C","D","E"])
y= np.array([3,8,5,10,6])



plt.barh(x,y , color="red")

plt.show()
```

```python
x= np.array(["A","B","C","D","E"])
y= np.array([3,8,5,10,6])



plt.bar(x,y , color="red", width=0.3)

plt.show()
```

## Scatter plot

```python
x=[1,2,3,4,5,6]
y=[4,3,8,5,2,7]
plt.title("Scatter plot")
plt.xlabel("month")
plt.ylabel("number")
plt.scatter(x,y, color="red")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[4,3,8,5,2,7]
plt.title("Scatter plot")
plt.xlabel("month")
plt.ylabel("number")
plt.scatter(x,y, color="red",s=100)
plt.show()
```
